```python
import pandas as pd
import requests
import json
```

```python
combined_data = pd.DataFrame()
file_paths = ["C:\CRUV\search_engine_api\search_api\services\data\input\college_degree.csv"]
```

```python
for file_path in file_paths:
    df = pd.read_csv(file_path)
    combined_data = pd.concat([combined_data, df], ignore_index=True)
```

```python
combined_data
```

```python
def seed_subjects(row):
    url = "http://127.0.0.1:8000/profile/save/degree"
    data = {
        "name": row["degree_title"],
        "education_level": row["degree_level"]
    }
    
    headers = {
        "Content-Type": "application/json"
    }

    response = requests.post(url, headers=headers, data=json.dumps(data))
    print(response.status_code)
    print(response.json())
```

```python
combined_data.apply(seed_subjects, axis=1)
```
